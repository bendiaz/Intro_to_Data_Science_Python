# Feature selection
## LASSO(least absolute shrinkage and selection operator) Regression

$\sum^{N}_{i=1} L^{i}(\beta) + \lambda \sum^M_{j=1}|\beta_j|^P$
$||y-X\beta||+\lambda||\beta||$


if P=2 -> Ridge (Elastic Net)
if P=1 -> LASSO

```python
#import fancyimpute
import seaborn as sns

iris_df = sns.load_dataset("iris")
iris_df
```

```python
import numpy as np
iris_X = iris_df.drop("species",1)

some_index = np.random.randint(0,iris_X.shape[0],100)
some_columns = np.random.randint(0,iris_X.shape[1],100)

for idx, col in zip(some_index,some_columns):
    iris_X.iloc[idx,col]=np.nan
iris_X
```

```python
from fancyimpute import KNN
import pandas as pd

pd.DataFrame(KNN(3).complete(iris_X),columns=iris_X.columns)
```
