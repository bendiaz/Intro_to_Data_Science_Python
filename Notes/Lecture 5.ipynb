

Descriptive statistic

The mean:
   $ \mu = E(x) = \sum_{x \exists \chi}  x P(x)$
   
   
 ${x_1,x_2,...,x_n} i.i.d.$
 
 
$M = \frac{1}{n} \sum_i=1 ^n \chi_i$



The variance:
    $\sigma^2 = E[(\chi - \mu)^2] = E(\chi^2) - E(\chi)^2$
    
    
$ S^2 = 1/(n-1) \sum_{i=1} ^n(\chi-M)^2$


$E[M] = E[\frac{1}{n} \sum_i=1 ^n \chi_i] = \frac{1}{n} \sum_{i=1} ^n E[\chi_i] = M = $


 $\frac{1}{n} \sum_{i=1} ^n \mu = \mu$
    


$E[XY] = E[X]E[Y]$ When independent


$ \prime{S^2} = \frac{1}{n} \sum_{i=1} ^n (Chi_i-M)^2 = \frac{1}{n} \sum_{i=1} ^n (\chi_i -\mu+\mu - M)^2$


   $ E[\prime{S^2}] = E[\frac{1}{n} \sum_{i=1} ^n (\chi - \mu)^2 + (\chi_i-M)^2 +2(\chi_i - \mu)(\mu - M)]$
    
    
   $ = \frac{1}{n} \sum_{i=1} ^n E[(\chi_i-M)^2] + E[\frac{1}{n} \sum_{i=1} ^n (\mu-M)^2 +2(\chi_i-\mu)(\mu-M) ]$
    
    
   $= \sigma^2 - 2E[(\mu - M)^2]+ E[(\mu-M)^2]$
    
    
   $ E[\prime{S^2}] = \sigma^2 - Var[M]$
    
    
  $  Var[M] = Var[1/n \sum_{i=1} ^n \chi_i] = \frac{1}{n^2}\sum_{i=1} ^n Var[\chi_i] = \frac{\sigma^2}{n}$
  
    
    
   $E[\prime{S^2}] = \sigma^2 - \sigma^2/n = \sigma*\frac{n-1}{n}$
    
    
   $ E[\frac{n}{n-1}\prime{S^2}]= \sigma^2$
    
    
  $ S^2 = \frac{1}{n-1}\sum_{i=1} ^n (\chi_i - M)^2$
    
    
 $ Var[a \chi] = a^2*Var[X] $
    
The Skewness: $\sigma = E[(\chi - \mu)^3 / \sigma^3$



The Kurtosis: $\kappa = E[(\chi - \mu)^4 / \sigma^4$

Dependency:

$P(x,y) \neq P(x)P(y)$



$P(x|y) \neq P(x)$

Pearson Correlation:

$\frac{E[(\chi-\mu_x)(\Upsilon - \mu_y)]}{\sigma_x \sigma_y} \rightarrow = $ covariance
$-1 \leq \rho_{x,y} \leq 1 $ -> Measure of linear dependency

Spearman's rank correlation:

$r_{x,y} = \rho_{rg_x,rg_y} = \frac{E[(r_{g_x}-\mu_{rg_x})(r_{g_y}-\mu_{rg_y})]}{\sigma_{rg_x}\sigma_{rgy}}$



$\chi = {1,6,2,9,0} => rg_x = {2,4,3,5,1} $


$sigma_{x,y} =1$

$r_{g_x} = {0,60,1,500,-8}$

$r_{g_y} = {2,4,3,5,1}$

Information Theory:



$H(x) = - \sum_{x \exists \chi} P(x) Log P(x) = E[-logP(x)]$


Mutual Information :

$ \leq I(x,y) = \sum_{x \exists \chi} \sum_{y \exists \upsilon} P(x,y) log(\frac{P(x,y)}{P(x)P(y)} \leq min(H(x),H(y))$

Jaccard Similarity Index:

$J(x,y) = \frac{I(x,y)}{H(x,y)}$

$H(x,y) = \sum_{x \exists \chi} \sum_{y \exists \upsilon}P(x,y)logP(x,y)$


```python
from sklearn.metrics.cluster import entropy
x = np.random.normal(size=1000)

entropy(x)
```

```python
def ent(data):
    unique, counts = np.unique(data,return_counts=True)
    p_data = counts/len(data)
    return stats.entropy(p_data)

ent(x)
```

```python
def ent1(data):
    unique, counts = np.unique(data,return_counts=True)
    p_data = counts/len(data)
    return -sum(p_data*np.log(p_data))

ent1(x)
```

```python
x = np.random.normal(size=1000)
entropy_df = pd.DataFrame(index=range(len(x)), columns=["Entropy"])

for i in range(len(x)):
    x[i] = 0 
    entropy_df.loc[i,"Entropy"] = ent1(x)
    
    #slowly making entropy go to 0
entropy_df.plot()
```
