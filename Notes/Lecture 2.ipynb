```python
def some_other_function(x):
    x += [2] #appending an element to the list
    return x

y = [3,4]
print(y)
print(some_other_function(y))
print(y)
```

```python
def last_function(x):
    temp_x = x.copy()
    temp_x +=[2] 
    return temp_x

y = [3,4]
print(y)
print(last_function(y))
print(y)
```
