```python
## TODO: write a function that create a data frame with stochastic paths
def run_simulation(num_particle, max_time, K, deltaT):
    temp_col = ['T_'+str(x) for x in range(max_time)]
    temp_DF = pd.DataFrame(np.sqrt(2*K*deltaT)*np.random.normal(size=(num_particle,max_time)), index=range(num_particle), columns=temp_col)
    temp_DF[['T_0']]=0.00
    temp_DF = temp_DF.cumsum(axis=1)
    #temp_DF.iloc[:,1:] = temp_DF.iloc[:,1:].cumsum(axis=1) + (np.sqrt(2*K*deltaT) * np.random.normal())
    #temp_DF.iloc[:,1:] = (np.sqrt(2*K*deltaT) * np.random.normal()) + (np.sqrt(2*K*deltaT) * np.random.normal(1))
    return temp_DF
    raise NotImplementedError

diffusion_filled_df = run_simulation(100, 1000, 2, 1)
diffusion_filled_df
```

```python
## TODO: create a dataframe `dispersion_df` by taking the square the elements
dispersion_df = diffusion_filled_df.pow(2)
# TODO: compute the mean and standard deviation of every column of the dispersion_df dataframe
dispersion_avg = dispersion_df.apply(lambda r: pd.Series({'mean': r.mean(), 'std': r.std()})).transpose()
dispersion_avg
```

```python
import scipy
from scipy.special import gamma as G

# TODO: write a function that compute g
def g(gamma, d):
    gcon = gamma**(4 / gamma)
    g1 = G((d + 2) / gamma)
    g2 = G(d / gamma)
    return (gcon * g1) / g2
    raise NotImplementedError
```

```python
def compute_prob_distribution(absolute_position, num_cuts):
    absolute_position_copy = pd.DataFrame(absolute_position.tolist(),columns=["Particles"])
    # TODO: cut absolute_position using qcut and num_cuts
    absolute_position_copy["cut"], bins = pd.qcut( absolute_position, q = num_cuts, retbins=True )
    
    # TODO: compute deltaR
    deltaR = bins[1:] - bins[:-1]
    
    # TODO: compute the counts per cuts of the column "cut"
    counts = absolute_position_copy.groupby('cut')['Particles'].count()
    
    # TODO: normalize the counts to get the probability distribution
    pdf = counts / (deltaR * len(absolute_position_copy["Particles"].tolist()))
    
    # TODO: compute the median per cuts of the column "cut" of the absolute position
    r = absolute_position_copy.groupby('cut')['Particles'].apply(np.median)
    
    # TODO_update: compute the median per cuts of the absolute position
    #absolute_position_copy.groupby('cut').indices.values().apply(np.median)
    
    return pdf, r

compute_prob_distribution(diffusion_filled_df3["T_999"].abs(), 50)
```

```python
# TODO: write a function that computes A
def A(gamma, d):
    a1 = G((d+2)/gamma)
    a2 = G(d/gamma)
    return ((a1/a2)**(gamma / 2))
# TODO: write a function that computes B
def B(gamma, d):
    b1 = (G((d + 2) / gamma)) ** (d/2)
    b2 = (G(1 / gamma)) ** ((d+2)/2)
    b = np.log(gamma*(G((d+2)/gamma)**(d/2))/(G(1/gamma)**((d+2)/2)))
    return b
# TODO: write a function that computes the dispersion
def dispersion(gamma, d, t, D):
    gdt = g(gamma, d)*(D*t)**(2 / gamma)
    return gdt
# TODO: write a function that computes the theoretical PDF
def pdf(gamma, d, r, t, D):
    pdf_front = (1/dispersion(gamma, d, t, D))**(d/2)
    pdf_back = np.exp(-A(gamma, d) * ((r/np.sqrt(dispersion(gamma, d, t, D)))**gamma) + B(gamma, d))
    return (pdf_front * pdf_back)
```

```python
# TODO: compute the estimated pdf and use the r array to show the theoretical one
est_pdf ,r = compute_prob_distribution(diffusion_filled_df3['T_999'].abs(),160)
est_pdf= est_pdf.tolist()
r = r.tolist()
theo_pdf = pdf(gamma=2, d=1,r=r,D=2, t=1000)

plt.figure()
plt.plot(est_pdf, label='Estimate')
plt.plot(theo_pdf, label="Theoretical")
plt.legend(loc="upper right")                                       
```

```python
## TODO:Plot the rescaled estimated pdf for t = 199, 399, 599, 799, 999
pdf1 ,r1 = compute_prob_distribution(diffusion_filled_df3['T_199'].abs(),160)
pdf2 ,r2 = compute_prob_distribution(diffusion_filled_df3['T_399'].abs(),160)
pdf3 ,r3 = compute_prob_distribution(diffusion_filled_df3['T_599'].abs(),160)
pdf4 ,r4 = compute_prob_distribution(diffusion_filled_df3['T_799'].abs(),160)
pdf5 ,r5 = compute_prob_distribution(diffusion_filled_df3['T_999'].abs(),160)


def x(r,gamma,t):
    disp = dispersion(gamma=gamma, d=1,t=t,D=2)
    value = (r/np.sqrt(disp))**gamma
    return value
    
def y(pdf,gamma, d, t, D,r):
    disp = dispersion(gamma=gamma,d=d,t=t,D=D)**(d/2)
    return np.log(pdf*disp)
    
x_1 = x(r=r1, gamma=2, t=199)
y_1 = y(pdf =pdf1,gamma=2, d=1,t=199, D=2, r=r1)
x_2 = x(r=r2, gamma=2, t=399)
y_2 = y(pdf=pdf2,gamma=2, d=1,t=399, D=2, r=r2)
x_3 = x(r=r3, gamma=2, t=599)
y_3 = y(pdf=pdf3,gamma=2, d=1,t=599, D=2, r=r3)
x_4 = x(r=r4, gamma=2, t=799)
y_4 = y(pdf=pdf4,gamma=2, d=1,t=799, D=2, r=r4)
x_5 = x(r=r5, gamma=2, t=999)
y_5 = y(pdf=pdf5,gamma=2, d=1,t=999, D=2, r=r5)

plt.figure()
plt.plot(x_1,y_1, label = "t=199")
plt.plot(x_2,y_2, label = "t=399")
plt.plot(x_3,y_3, label = "t=599")
plt.plot(x_4,y_4, label = "t=799")
plt.plot(x_5,y_5, label = "t=999")
plt.rc("text",usetex=True)
plt.ylabel(r"$log(P(r,t)(r^{2}(t)^{\frac{1}{2}})$", fontsize=25)
plt.xlabel(r"$(\frac{r}{\sqrt{ \langle r^2(t) \rangle}})^{\gamma}$", fontsize=25)
plt.legend(loc="upper right")

```
